```python
import os
import torch
import pandas as pd
import cv2
```

```python
def searchandgenerate(img_dir,res_path,model):
    
    

    if os.path.isdir(res_path) == False:
        os.mkdir(res_path)
        
    
        
    imgs_names = os.listdir(img_dir)
    
    
    for directory in imgs_names:
        if directory.endswith('.png'):
            generatecsv(img_dir+'/'+directory,res_path+'/'+directory,model)
            
        if os.path.isdir(img_dir+'/'+directory):
            searchandgenerate(img_dir+'/'+directory,res_path+'/'+directory,model)
           

            
def generatecsv(img_dir,res_path,model):
    width = cv2.imread(img_dir).shape[1] #assuming all images have identical dimensions
    height = cv2.imread(img_dir).shape[0]
    results_to_csv = pd.DataFrame(columns=["x1","y1","x2","y2","x3","y3","x4","y4","xc","yc","object"])
    results=model(img_dir)

    for p in range(0,len(results.pandas().xyxy[0])):
        results_to_csv = results_to_csv.append(pd.DataFrame({"x1":[results.pandas().xyxy[0].at[p,"xmin"]],"y1":[results.pandas().xyxy[0].at[p,"ymin"]],"x2":[results.pandas().xyxy[0].at[p,"xmin"]],"y2":[results.pandas().xyxy[0].at[p,"ymax"]],"x3":[results.pandas().xyxy[0].at[p,"xmax"]],"y3":[results.pandas().xyxy[0].at[p,"ymin"]],"x4":[results.pandas().xyxy[0].at[p,"xmax"]],"y4":[results.pandas().xyxy[0].at[p,"ymax"]],"xc":[0],"yc":[0],"object":[results.pandas().xyxy[0].at[p,"name"]]},index=[img_dir]))
    results_to_csv.loc[img_dir,"xc"]=(results_to_csv.loc[img_dir,"x1"]+results_to_csv.loc[img_dir,"x3"])/2
    results_to_csv.loc[img_dir,"yc"]=(results_to_csv.loc[img_dir,"y1"]+results_to_csv.loc[img_dir,"y2"])/2
    results_to_csv.x1 = results_to_csv.x1/width
    results_to_csv.x2 = results_to_csv.x2/width
    results_to_csv.x3 = results_to_csv.x3/width
    results_to_csv.x4 = results_to_csv.x4/width
    results_to_csv.xc = results_to_csv.xc/width
    results_to_csv.y1 = results_to_csv.y1/height
    results_to_csv.y2 = results_to_csv.y2/height
    results_to_csv.y3 = results_to_csv.y3/height
    results_to_csv.y4 = results_to_csv.y4/height
    results_to_csv.yc = results_to_csv.yc/height
    results_to_csv.to_csv(path_or_buf = res_path + "_normalized.csv",index=False,header=False )
    
```

```python
for img_name in imgs_names[0:40]:
    image=cv2.imread(img_dir+'/'+img_name)
    res=pd.read_csv(img_dir+"/results/"+img_name+".csv",index_col=False,names=["x1","y1","x2","y2","x3","y3","x4","y4","xc","yc","object"])
    for p in range(0,len(res)):
        start_point=(int(res.at[p,"x1"]),int(res.at[p,"y1"]))
        end_point=(int(res.at[p,"x4"]),int(res.at[p,"y4"]))
        image=cv2.rectangle(image,start_point,end_point,(0,255,0),2)
    cv2.imshow(img_name,image)
    cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
